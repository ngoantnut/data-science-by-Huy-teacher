```python
he3d=plt.axes(projection='3d') # Thc hin khai bo 1 h ta  3D
# VD v ng trong khng gian 3D
zline=np.linspace(0,15,1000) # Khi to 1000 con s trong khong t 0-15
xline=np.sin(zline)
yline=np.cos(zline)
#print(zline.shape,x)
# V biu 
he3d.plot3D(xline,yline,zline)
```

* Bc 2: Dng d liu cho 3 trc x,y,z
* Bc 3: V dng hm plot3D(x,y,z)

```python
# nh ngha hm z=f(x,y), hm ny ch c tnh cht v d  sinh d liu
# Trong thc t cc d liu 3 trc x,y,z l c sn, khng cn sinh t ng
def f(x,y):
    return np.sin(np.sqrt(x**2 + y**2))  
# Sinh ngu nhin d liu trc x v y
#x= np.linspace(-6,6,30)
x=np.random.randint(0,100,50)
y=np.random.randint(200,250,50)
#print(x)
#y= np.linspace(-6,6,30)
# To ra cc li trong khng gian c
X,Y = np.meshgrid(x,y)
# tnh gi tr ca Z trong mi im ca khng gian 3D
Z=f(X,Y)

# v dng contour3D
he3d2=plt.axes(projection='3d')
he3d2.contour3D(X,Y,Z,10)
#plt.show()
```
